```python
from openpyxl import load_workbook
read_wb =load_workbook("./청와대 청원데이터.xlsx")
read_ws=read_wb.active
```

```python
list_excel=[]
for i in range (1,151):
    print(read_ws.cell(i,1).value)
    list_excel.append(read_ws.cell(i,1).value)
```

```python
list_excel
```

```python
from konlpy.tag import Kkma
import collections
```

```python
kkma=Kkma()
```

```python
list_temp=[]
```

```python
for row in list_excel:
    print(kkma.nouns(row))
    list_temp=list_temp+kkma.nouns(row)
```

```python
list_temp
```

```python
result_list=[]
```

```python
for check in list_temp:
    if len(check)>1:
        result_list.append(check)
```

```python
result_list
```

```python
collections.Counter(result_list)
```

```python
collections.Counter(result_list).most_common(20)
```

```python
from PIL import Image
from wordcloud import STOPWORDS

mask=np.array(Image.open("cloud.png"))
stopwards=set(STOPWORDS)
stopwards.add("said")
```

```python
mask
```

```python
wc=WordCloud(background_color="white",max_words=2000,mask=mask,stopwords=stopwards,font_path = 'C:/Windows/Fonts/malgun.ttf')
wc=wc.generate(last_text)
wc.words_
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc,interpolation="bilinear")
plt.axis("off")
plt.show()
```
